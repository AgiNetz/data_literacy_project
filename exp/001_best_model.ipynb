```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import logging
logging.root.setLevel(10)

import sys
import json
import os
sys.path.insert(0, os.path.realpath('..'))
os.chdir('..')

from src import data
from src.model import get_splits

with open('config.json', 'r') as cfg:
    config = json.load(cfg)
    
random_seed = 42
```

```python
from src.model import find_best_linear_model, find_best_ridge_model, find_best_neural_model

def get_samples_variation(categories, expenditure_unit):
    top = True if categories=='parent' else False
    samples_index = 0 if expenditure_unit=='pcnt' else 1
    dataset = data.load_data(config, top)
    clean_data = data.filter_bad_data(dataset)
    samples = data.create_samples(clean_data)
    return get_splits(samples[samples_index], test_size=config["test_size"], random_seed=random_seed)

categories = ['parent', 'child']
units = ['pcnt', 'usd']

best = {"pcnt": {"parent": {}, "child": {}}, 
        "usd": {"parent": {}, "child": {}}}
for cat in categories:
    for unit in units:
            train, test = get_samples_variation(cat, unit)
            train['Total expenditure per capita (1000s USD)'] /= 1000
            fts = train.iloc[:,2:-2].to_numpy()
            labels = train["Happiness score"].to_numpy()
            best[unit][cat]["linear"] = find_best_linear_model(fts, labels)
            best[unit][cat]["ridge"] = find_best_ridge_model(fts, labels)
            best[unit][cat]["neural"] = find_best_neural_model(fts, labels, random_state=random_seed)
            logging.info("Finished variation {c} {u}".format(c=cat, u=unit))
        
print(best)
        
    
```
